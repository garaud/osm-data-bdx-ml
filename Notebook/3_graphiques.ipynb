### Plot scatter on PCA values

```python
def plot_scatter(data):
    """
    
    """
    
    init_notebook_mode(connected=True)
    
    
    cluster = data.reset_index()
    cluster_id_list = cluster["Xclust"].unique().tolist()
    cluster_id_list.sort()
    
    # Init list of trace
    data_graph = []
    
    for cluster_id in cluster_id_list:
        trace = go.Scatter3d(x=cluster[cluster['Xclust'] == cluster_id]['PC1'],
                             y=cluster[cluster['Xclust'] == cluster_id]['PC2'],
                             z=cluster[cluster['Xclust'] == cluster_id]['PC3'],
                               text = cluster[cluster['Xclust'] == cluster_id]['uid'],
                               name = "Cluster " + str(cluster_id),
                               mode="markers",
                               marker={
                                   'size': 5,
                                   #'opacity': 0.7,
                                   'line': {'width': 0.1, 'color': 'white'}
                                })
        data_graph.append(trace)
        
    # Design graph
    layout = dict(
        title='Plot PCA ',
        xaxis=dict(title='PC1'),
        yaxis=dict(title='PC2'),
    )

    fig = dict(data=data_graph, layout=layout)
    iplot(fig)
```

```python
plot_scatter(cluster)
```
